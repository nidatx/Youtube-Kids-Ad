Aggregated


```python
agg_high1 = {}
for key,value in high1.items():
  if (key=="country"):
    agg_high1["region"] = ["high"]
  else:
    agg_high1[key] = [sum(value)]

agg_low1 = {}
for key,value in low1.items():
  if (key=="country"):
    agg_low1["region"] = ["low"]
  else:
    agg_low1[key] = [sum(value)]

```

```python
agg_df1 = pd.concat([pd.DataFrame(agg_high1),pd.DataFrame(agg_low1)],ignore_index=True)
```

```python
agg_df1
```

```python
# Melt the DataFrame to convert columns to rows
melted_agg1 = pd.melt(agg_df1, id_vars=["region"], var_name="category", value_name="value")
melted_agg1
```
